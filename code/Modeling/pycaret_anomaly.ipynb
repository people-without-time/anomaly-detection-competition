```python
import numpy as np
import pandas as pd
import random

from pathlib import Path
from pycaret.anomaly import *
```

```python
RANDOM_SEED = 42
DATA_PATH = Path("../../data")

np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
```

```python
train_data = pd.read_csv(DATA_PATH / "train.csv")
test_data = pd.read_csv(DATA_PATH / "test.csv")
```

```python
exp_name = setup(train_data, ignore_features=["faultNumber", "simulationRun", "sample"],session_id=42)
```

```python
models()
```

```python
method = ['abod', 'cluster', 'iforest', 'histogram', 'knn', 'lof', 'svm', 'pca', 'mcd', 'sod']

test_pred = np.zeros(test_data.shape[0],)
preds = []
```

```python
from tqdm.auto import tqdm
for id in tqdm(method):
    try :
        # 모델 생성
        model = create_model(id, fraction=0.31)
        
        # 모델을 통해 Test 데이터를 pred
        pred_df = predict_model(model, test_data.iloc[:, 2:])
        
        # 결과 출력
        print(f"== model = {id} ==")
        print(pred_df["Anomaly"].value_counts())
        
        # 예측값 추합
        pred = pred_df["Anomaly"]
        preds.append(pred.tolist())
        test_pred += pred
    except :
        pass
```
